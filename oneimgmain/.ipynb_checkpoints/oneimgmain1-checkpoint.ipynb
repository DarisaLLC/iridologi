```python
import os
import cv2 as cv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def auto_canny(image, sigma=0.33):
    # compute the median of the single channel pixel intensities
    v = np.median(image)
    
    # apply automatic Canny edge detection using computed median
    lower = int(max(0, (1.0-sigma) * v))
    upper = int(min(255, (1.0+sigma) * v))
    canny = cv.Canny(image, lower, upper)
    
    return canny
```

```python
def rotate(image, angle):
    height, width = image.shape[:2]
    rot_mat = cv.getRotationMatrix2D((width/2, height/2), angle, 1)
    rotated_img = cv.warpAffine(image, rot_mat, (width,height))
    return rotated_img
```

```python
def display(image, title):
    plt.title(title)
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    plt.show()
```

```python
def display_hist(image, title):
    plt.figure()
    plt.title(title)
    plt.hist(image.ravel(),256,[0,256])
    plt.show()
```

```python
image = cv.imread('../Data_ext/ada/ada_050.JPG')
print(image.shape)
display(image,'image')

b,g,r = cv.split(image)
pdb = pd.DataFrame(b)
print('Blue: baris(250:255), kolom(100:105)')
print(np.array(pdb.iloc[250:255, 100:105]))

pdg = pd.DataFrame(g)
print('\nGreen: baris(250:255), kolom(100:105)')
print(np.array(pdg.iloc[250:255, 100:105]))

pdr = pd.DataFrame(r)
print('\nRed: baris(250:255), kolom(100:105)')
print(np.array(pdr.iloc[250:255, 100:105]))

pd.DataFrame(np.asarray(b)).to_csv('./blue.csv',header=False, index=False)
pd.DataFrame(np.asarray(g)).to_csv('./green.csv',header=False, index=False)
pd.DataFrame(np.asarray(r)).to_csv('./red.csv',header=False, index=False)
```

```python
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
print(gray.shape)
display(gray, 'gray')

pdgr = pd.DataFrame(gray)
print('Gray: baris(250:255), kolom(100:105)')
print(np.array(pdgr.iloc[250:255, 100:105]))

pd.DataFrame(np.asarray(gray)).to_csv('./grayscale.csv',header=False, index=False)

display_hist(gray,'gray')
```

```python
clahe = cv.createCLAHE(clipLimit=5.0, tileGridSize = (5,5))
clahe_img = clahe.apply(gray)
display(clahe_img, 'clahe hist')

pd.DataFrame(np.asarray(clahe_img)).to_csv('./hist_clahe.csv',header=False, index=False)

display_hist(clahe_img, 'clahe_hist')
```

```python
blur_clahe = cv.medianBlur(clahe_img, 5)
display(blur_clahe, 'blur_clahe')

pd.DataFrame(np.asarray(blur_clahe)).to_csv('./medianBlur.csv',header=False, index=False)

# display_hist(blur_clahe, 'blur_clahe')
```

```python
img = cv.cvtColor(blur_clahe.copy(), cv.COLOR_GRAY2RGB)

_, thresh = cv.threshold(blur_clahe, 10, 255, cv.THRESH_BINARY_INV)
display(thresh, 'thresh')
    
contours, hierarchy = cv.findContours(thresh, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
maxContour = 0
for contour in contours:
    contourSize = cv.contourArea(contour)
    if contourSize > maxContour:
        maxContour = contourSize
        maxContourData = contour
# Find enclosing circle to pupil
(x,y), r = cv.minEnclosingCircle(maxContourData)
center = (int(x), int(y))
radius = int(r)
masked_pupil = cv.circle(img, center, radius+10, (255,255,255), -1)
display(img, 'masked_pupil')
```

```python
img = cv.cvtColor(masked_pupil.copy(), cv.COLOR_RGB2GRAY)
iris_img = masked_pupil.copy()

circles = cv.HoughCircles(img, cv.HOUGH_GRADIENT,1, int(image.shape[0]), param1=50, param2=50)
if circles is not None:
    circles = np.uint16(np.around(circles))
    for (x,y,r) in circles[0,:]:
        iris_center = (int(x), int(y))
        iris_radius = int(r)
        iris = cv.circle(iris_img, iris_center, iris_radius, (0,255,0),5)
        iris = cv.circle(iris_img, iris_center, 2, (0,255,0), 5)

display(iris, 'iris')
```

```python
# Convert Iris Circle into Cartesian Image
cart = cv.linearPolar(img, iris_center, iris_radius, cv.WARP_FILL_OUTLIERS)
rot_cart = rotate(cart, -90)
display(cart, 'cartesian')
display(rot_cart, 'rotated_cart')
```

```python
[y,x] = rot_cart.shape

img = cv.cvtColor(rot_cart.copy(), cv.COLOR_GRAY2RGB)
target_box = cv.rectangle(img, (0,0), (int(x/12),int(y)), (0,255,0), 5)
display(target_box,'target_box')

target = rot_cart[0:int(y), 1:int(x/12)]
display(target,'target')
```

```python
img = cv.cvtColor(target.copy(), cv.COLOR_GRAY2RGB)
_, thresh = cv.threshold(target, 250, 255, cv.THRESH_BINARY)
contours, _ = cv.findContours(thresh, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
maxContour = 0
for contour in contours:
    contourSize = cv.contourArea(contour)
    if contourSize > maxContour:
        maxContour = contourSize
        maxContourData = contour
rect = cv.boundingRect(maxContourData)
x,y,w,h = rect
pupil = cv.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)
display(pupil,'pupil_region')
```

```python
crop_pupil = target[y+h+5:, 0:]
display(crop_pupil,'cropped_pupil')
```

```python
img = cv.cvtColor(crop_pupil.copy(), cv.COLOR_GRAY2RGB)
[y,x] = crop_pupil.shape
target_roi = cv.rectangle(img, (0,0), (x, int(y/2)), (0,255,0), 2)
display(target_roi,'target_roi')

roi = crop_pupil[0:int(y/2), 0:]
display(roi,'roi')
```

```python
roi_res = cv.resize(roi, (50,50))
display(roi_res, 'roi_res')

# canny = cv.Canny(roi_res, 50, 100)
# display(canny, 'canny')

autocanny = auto_canny(roi_res)
display(autocanny, 'autocanny')
```

```python
from csv import reader
def readCSV(filename):
    dataset = list()
    with open(filename, 'r') as file:
        csv_reader = reader(file)
        for row in csv_reader:
            if not row:
                continue
            dataset.append(row)
    return dataset

dataset = readCSV('../features_df.csv')
dataset = pd.DataFrame(dataset)
print(np.array(dataset).shape)
print(np.array(dataset))

print('\n')
print(np.array(dataset.iloc[50:55, 50:55]))
```
